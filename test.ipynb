```python
df = df.drop(columns=['Unnamed: 0','Job Description', 'Founded', 'Revenue', 'Competitors', 'Easy Apply'])
df.head()
```

```python
unique = df["Location"].unique()

list1 = unique.tolist()
print(list1)
```

```python
from geopy import geocoders
gn = geocoders.GeoNames(username='angheljf18')
my_dict = {"Latitude":[], "Longitude": []}
# print(gn.geocode("Cleveland, OH").latitude, gn.geocode("Cleveland, OH").longitude)
# my_dict["City"].append(list1)
for i in list1:
#     print(gn.geocode(i).latitude, gn.geocode(i).longitude)
    my_dict["Latitude"].append(gn.geocode(i).latitude)
    my_dict["Longitude"].append(gn.geocode(i).longitude)
```

```python
for i in list1:
    my_dict["Location"]=list1
```

```python
my_dict
```

```python
lat_df = pd.DataFrame.from_dict(my_dict)
lat_df.head()
```

```python
new_df = pd.merge(df, lat_df, how='left', on='Location')
new_df
```

```python
new_df.to_csv('test.csv', index=False)
```
